```python
import numpy as np
import pandas as pd
from pathlib import Path
import joblib
from sklearn.metrics import accuracy_score, classification_report

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    roc_auc_score,
)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from xgboost import XGBClassifier, XGBRegressor



X_train = pd.read_parquet("../data/processed/X_train.parquet")
X_test  = pd.read_parquet("../data/processed/X_test.parquet")



def load_label_csv(path: str) -> pd.Series:
    df = pd.read_csv(path, header=None)

    
    if df.shape[1] >= 2:
        s = df.iloc[:, 1]
    else:
        s = df.iloc[:, 0]

    s = pd.to_numeric(s, errors="coerce")  
    return s



y_train_risk = load_label_csv("../data/processed/y_train_risk.csv")
y_test_risk  = load_label_csv("../data/processed/y_test_risk.csv")

y_train_fatigue = load_label_csv("../data/processed/y_train_fatigue.csv")
y_test_fatigue  = load_label_csv("../data/processed/y_test_fatigue.csv")



print("Before cleaning:")
print("X_train:", X_train.shape, "y_train_risk:", y_train_risk.shape, "y_train_fatigue:", y_train_fatigue.shape)
print("X_test :", X_test.shape,  "y_test_risk :", y_test_risk.shape,  "y_test_fatigue :", y_test_fatigue.shape)
print("NaNs in y_train_risk:", int(y_train_risk.isna().sum()))
print("NaNs in y_test_risk :", int(y_test_risk.isna().sum()))


y_train_risk = y_train_risk.iloc[:len(X_train)]
y_train_fatigue = y_train_fatigue.iloc[:len(X_train)]
y_test_risk = y_test_risk.iloc[:len(X_test)]
y_test_fatigue = y_test_fatigue.iloc[:len(X_test)]


train_mask = y_train_risk.notna()
test_mask  = y_test_risk.notna()

X_train = X_train.loc[train_mask.values].copy()
y_train_risk = y_train_risk.loc[train_mask.values].astype(int)
y_train_fatigue = y_train_fatigue.loc[train_mask.values].astype(float)

X_test = X_test.loc[test_mask.values].copy()
y_test_risk = y_test_risk.loc[test_mask.values].astype(int)
y_test_fatigue = y_test_fatigue.loc[test_mask.values].astype(float)

print("\nAfter cleaning:")
print("X_train:", X_train.shape, "y_train_risk:", y_train_risk.shape, "y_train_fatigue:", y_train_fatigue.shape)
print("X_test :", X_test.shape,  "y_test_risk :", y_test_risk.shape,  "y_test_fatigue :", y_test_fatigue.shape)


assert len(X_train) == len(y_train_risk) == len(y_train_fatigue), "Train lengths still mismatch!"
assert len(X_test) == len(y_test_risk) == len(y_test_fatigue), "Test lengths still mismatch!"



feature_cols = list(X_train.columns)
label_classes = ["Low", "Medium", "High"]



risk_model = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("model", XGBClassifier(
        n_estimators=600,
        max_depth=5,
        learning_rate=0.05,
        subsample=0.9,
        colsample_bytree=0.9,
        objective="multi:softprob",
        num_class=3,
        eval_metric="mlogloss",
        random_state=42
    ))
])

risk_model.fit(X_train, y_train_risk)

pred = risk_model.predict(X_test)
proba = risk_model.predict_proba(X_test)

print("\n=== Risk Model (XGBClassifier) ===")
print("Accuracy:", accuracy_score(y_test_risk, pred))
print("Confusion Matrix:\n", confusion_matrix(y_test_risk, pred))
print("Classification Report:\n", classification_report(y_test_risk, pred))
print("ROC-AUC OvR:", roc_auc_score(y_test_risk, proba, multi_class="ovr"))


def fatigue_to_risk(f):
    if f < 35:
        return 0
    elif f < 65:
        return 1
    else:
        return 2

baseline_pred = np.array([fatigue_to_risk(f) for f in y_test_fatigue])

print("Rule-baseline Accuracy:", accuracy_score(y_test_risk, baseline_pred))
print(classification_report(y_test_risk, baseline_pred))


fatigue_model = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("model", XGBRegressor(
        n_estimators=600,
        max_depth=4,
        learning_rate=0.05,
        subsample=0.9,
        colsample_bytree=0.9,
        objective="reg:squarederror",
        random_state=42
    ))
])

fatigue_model.fit(X_train, y_train_fatigue)
fat_pred = fatigue_model.predict(X_test)

mae = mean_absolute_error(y_test_fatigue, fat_pred)
rmse = np.sqrt(mean_squared_error(y_test_fatigue, fat_pred))
r2 = r2_score(y_test_fatigue, fat_pred)

print("\n=== Fatigue Model (XGBRegressor) ===")
print("Fatigue MAE:", mae)
print("Fatigue RMSE:", rmse)
print("Fatigue R2:", r2)



OUT = Path("../model_store")
OUT.mkdir(parents=True, exist_ok=True)

joblib.dump(risk_model, OUT / "risk_model.pkl")
joblib.dump(fatigue_model, OUT / "fatigue_model.pkl")
joblib.dump(feature_cols, OUT / "feature_cols.pkl")
joblib.dump(label_classes, OUT / "risk_label_classes.pkl")

print("\nSaved final models to model_store/")
print("Saved files:", [p.name for p in OUT.glob("*.pkl")])
```
