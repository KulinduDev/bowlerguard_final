```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

DATA_PATH = Path("../data/raw/bowlerguard_5000_v3.csv")  
OUT_PATH  = Path("../data/processed/clean.parquet")
```

```python
df = pd.read_csv(DATA_PATH)

print("Rows:", len(df))
print("Columns:", len(df.columns))
df.head()
```

```python

pd.DataFrame({
    "column": df.columns,
    "dtype": df.dtypes.astype(str)
})
```

```python

missing = df.isna().mean().sort_values(ascending=False)
dupes = df.duplicated().sum()

print("Duplicate rows:", dupes)
missing.head(20)
```

```python

df["match_date"] = pd.to_datetime(df["match_date"], errors="coerce")
df["match_start_date"] = pd.to_datetime(df["match_start_date"], errors="coerce")

print("Date parsing failures (match_date):", df["match_date"].isna().sum())
print("Date parsing failures (match_start_date):", df["match_start_date"].isna().sum())

print("Min date:", df["match_date"].min())
print("Max date:", df["match_date"].max())
```

```python


checks = {
    "deliveries_match_min": df["deliveries_match"].min(),
    "deliveries_match_max": df["deliveries_match"].max(),
    "overs_match_min": df["overs_match"].min(),
    "overs_match_max": df["overs_match"].max(),
    "fielding_min": df["inferred_fielding_time_minutes"].min(),
    "fielding_max": df["inferred_fielding_time_minutes"].max(),
    "temp_min": df["avg_temp_c"].min(),
    "temp_max": df["avg_temp_c"].max(),
    "humidity_min": df["avg_humidity_pct"].min(),
    "humidity_max": df["avg_humidity_pct"].max(),
    "esi_min": df["esi_norm"].min(),
    "esi_max": df["esi_norm"].max(),
    "acwr_std_min": df["acwr_std"].min(),
    "acwr_std_max": df["acwr_std"].max(),
    "age_min": df["age_years"].min(),
    "age_max": df["age_years"].max(),
}
pd.Series(checks).to_frame("value")
```

```python

df["injury_risk_label"].value_counts(dropna=False)
```

```python

plt.figure(figsize=(6,4))
df["acwr_std"].hist(bins=30)
plt.title("Distribution of ACWR (standard)")
plt.xlabel("acwr_std")
plt.ylabel("count")
plt.show()

plt.figure(figsize=(6,4))
df["esi_norm"].hist(bins=30)
plt.title("Distribution of Environmental Stress Index (esi_norm)")
plt.xlabel("esi_norm")
plt.ylabel("count")
plt.show()
```

```python


df_clean = df.drop_duplicates().copy()

df_clean = df_clean.dropna(subset=["match_date", "bowler", "injury_risk_label"])

df_clean["match_day"] = df_clean["match_day"].astype(int)
df_clean = df_clean[(df_clean["match_day"] >= 1) & (df_clean["match_day"] <= 5)]

print("Clean rows:", len(df_clean))
df_clean.head()
```

```python

df_clean.to_parquet(OUT_PATH, index=False)
print("Saved:", OUT_PATH.resolve())
```
